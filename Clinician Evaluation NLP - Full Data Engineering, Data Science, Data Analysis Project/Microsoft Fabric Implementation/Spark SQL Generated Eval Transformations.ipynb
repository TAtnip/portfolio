# Spark SQL Generated Evaluation Transormations for Business Logic and PowerBI Analysis

```python
%%sql
--Helpful changes for easier to read analysis
ALTER TABLE pt_eval_generated
ADD COLUMN tx_prior_loc STRING;

UPDATE pt_eval_generated
SET tx_prior_loc = CASE 
    WHEN prior_loc = 0 THEN 'Ranch Home'
    WHEN prior_loc = 1 THEN 'Two Story Home'
    WHEN prior_loc = 2 THEN 'Apartment'
    WHEN prior_loc = 3 THEN 'Independent Living Facility'
    WHEN prior_loc = 4 THEN 'Assisted Living Facility'
    WHEN prior_loc = 5 THEN 'Skilled Nursing Facility'
    WHEN prior_loc = 6 THEN 'Inpatient Rehabilitation Hospital'
    WHEN prior_loc = 7 THEN 'Long-term Care Facility'
END;
```

```python
%%sql
--Helpful changes for easier to read analysis
ALTER TABLE pt_eval_generated
ADD COLUMN tx_need_rehab STRING;

UPDATE pt_eval_generated
SET tx_need_rehab = CASE
    WHEN need_rehab = 0 THEN 'Home'
    WHEN need_rehab = 1 THEN 'Rehab'
END;
```

```python
%%sql 
--Analysis of strictly those patients from home for PowerBI
CREATE OR REPLACE TABLE pt_eval_home AS 
(SELECT * FROM pt_eval_generated
WHERE prior_loc IN (0, 1, 2 , 3, 4));
```

```python
%%sql 
--Proportions for PowerBI
CREATE OR REPLACE TABLE rehab_proportions AS
(SELECT tx_prior_loc, AVG(los) as average_los, ROUND((SUM(CASE WHEN need_rehab = 1 THEN 1 ELSE 0 END) *1.0 /COUNT(*)) * 100,2) AS RehabProportion 
FROM pt_eval_generated
GROUP BY tx_prior_loc
ORDER BY tx_prior_loc);
```

```python
%%sql

ALTER TABLE rehab_proportions
ADD COLUMN sort_order int;
UPDATE rehab_proportions SET sort_order = 1 WHERE tx_prior_loc = 'Ranch home';
UPDATE rehab_proportions SET sort_order = 2 WHERE tx_prior_loc = 'Two story home';
UPDATE rehab_proportions SET sort_order = 3 WHERE tx_prior_loc = 'Apartment';
UPDATE rehab_proportions SET sort_order = 4 WHERE tx_prior_loc = 'Independent Living Facility';
UPDATE rehab_proportions SET sort_order = 5 WHERE tx_prior_loc = 'Assisted Living Facility';
UPDATE rehab_proportions SET sort_order = 6 WHERE tx_prior_loc = 'Skilled Nursing Facility';
UPDATE rehab_proportions SET sort_order = 7 WHERE tx_prior_loc = 'Inpatient Rehabilitation Hospital';
UPDATE rehab_proportions SET sort_order = 8 WHERE tx_prior_loc = 'Long-term Care Facility';
```
